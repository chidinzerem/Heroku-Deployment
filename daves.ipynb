```python
from sklearn.datasets import fetch_openml
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
```

```python
mice = fetch_openml(name='miceprotein', version=4)
```

```python
mice.data
```

```python
mice.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(mice.data, mice.target, random_state=0)

pipe = Pipeline([('impute', SimpleImputer()), ('scaler', StandardScaler()), ('scv', SVC())])
pipe.fit(X_train, y_train)
```

```python
import pickle

with open('pipe.pickle', 'wb') as f:
    pickle.dump(pipe, f)
```

```python
import json

with open('testdata.json', 'w') as f:
    json.dump(mice.data.iloc[0].values.tolist(), f)
```
